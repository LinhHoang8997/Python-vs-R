# Preparing data for K-means clustering

Limit the set to only 3 features for easier demonstration.

```r
df <- df %>% select(-StoreID,-TotalSold, -Original)
# Deselect Total Sold (because of covariance) and StoreID (irrelevance)
```

```r
summary(df)
```

The range and scale of each features differ greatly, so we must scale the features appropriately.

We can use the Standard Scale for this purpose.
We observe from the descriptive stats that there might be outliers, so let get them remove. If not much was removed from the set, then no need to worry.  

```r
# Standardized variables (Z-score) to prepare for clustering. Otherwise, the result may be skewed due to the scale of each variable.
df <- as.data.frame(scale(df))
head(df,3)
```

```r
df_clean <- df
```

# Decide how many K-means clusters to use

```r
# Use var() to calculate variance
# Fill empty list wss with the first value - the Total Variance for when there is no cluster.
wss <- (nrow(df_clean)-1)*sum(apply(df_clean,2,var))
```

```r
# Set max clusters arbitrarily at 15.
max_clusters = 30
```

```r
#Find the optimal number of centroids
# Calculate the within groups SSE (Cohesion) for different numbers of cluster solutions
set.seed(1)

for (i in 2:max_clusters) 
    wss[i] <- sum(kmeans(df_clean, centers=i, iter.max=500)$withinss)
options(repr.plot.width = 6, repr.plot.height = 4)
plot(1:max_clusters, wss, type="b", xlab="Number of Clusters",
  ylab="Within groups sum of squares",cex.lab=0.75,cex.axis=0.75)
```

Based on where the line "breaks", we arbitrarily choose 10 according to the Elbow method.



R provides Lloyd's algorithm as an option for `kmeans()`; the default algorithm, by Hartigan and Wong (1979) is smarter. 

"Like MacQueen's algorithm (MacQueen, 1967), it updates the centroids any time a point is moved; it also makes clever (time-saving) choices in checking for the closest cluster. On the other hand Lloyd's k-means algorithm is the first and simplest of all these clustering algorithms."

**From** https://stackoverflow.com/a/20450717/8750096

# Run K-means clustering

```r
# Settle on one cluster for demonstration
kM_model <- kmeans(df_clean, centers=6, iter.max=500, nstart = 25)
```

**Note:**

Compared to Python, R's `kmeans` package provides much more details regarding the models and its components.

This is important to our persona - The Beginner.

```r
kM_model
```

# Visualizing the results

## In 2D (only two features)

```r
df_clean %>%
  as_tibble() %>%
  mutate(cluster = kM_model$cluster) %>%
  ggplot(aes(Fashion, Stretch, color = factor(cluster))) +
  geom_point(shape="o", size=5)

# Mutate adds a column with the name "cluster" and fill it with the values in kM_model$cluster
```
